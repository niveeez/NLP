### N-Grams

```python
import pandas as pd
```

```python
messages = pd.read_csv('SMSSpamCollection.csv', sep='\t',names=['label', 'message'])
```

```python
import nltk
import re
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
```

Using Lemmatization

```python
from nltk.stem import WordNetLemmatizer
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
corpus = []
for i in range(0,len(messages)):    
    review = re.sub('[^a-zA-Z]', ' ', messages['message'][i]) #Remove special characters
    review = review.lower() #Change to lower case
    review = review.split()
    review = [lemmatizer.lemmatize(lines) for lines in review if lines not in stopwords.words('english')] #Removing stopwords and stemming
    review = ' '.join(review)
    corpus.append(review)
```

```python
corpus
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

Take top most occuring words using max_features

```python
tfidf = TfidfVectorizer(max_features=100)
```

```python
X = tfidf.fit_transform(corpus).toarray()
```

```python
X
```

```python
exp = pd.DataFrame(X)
```

```python
exp.to_csv()
```

With ngrams

```python
tfidf = TfidfVectorizer(max_features=100, ngram_range=(2,2))
X = tfidf.fit_transform(corpus).toarray()
```
